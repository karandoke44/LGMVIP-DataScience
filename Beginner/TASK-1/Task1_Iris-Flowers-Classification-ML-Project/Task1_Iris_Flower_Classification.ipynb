# Building Model , Training and Testing

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Species_Flower'])
Y = df['Species_Flower']
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size = 0.3, random_state=2022)
```

```python
print("x_train: ", len(x_train))
print("x_test: ", len(x_test))
print("y_train: ", len(y_train))
print("y_test: ", len(y_test))
```

# 1. Logistic Regression Algorithm

```python
# Initialize a Logistic Regression
lg= LogisticRegression(max_iter=1000)
lg.fit(x_train,y_train)
```

```python
# Predict on the test set and calculate accuracy
y_pred = lg.predict(x_test)
print("Pridicted values on Test Data", y_pred)
```

```python
score=accuracy_score(y_test,y_pred)
```

```python
print('Logistic Regression')
print(classification_report(y_test,y_pred))
print(f'Accuracy: {round(score*100,2)}%')
conf_mat = confusion_matrix(y_test, y_pred)
axes = plt.axes()
conf_mat_df = conf_mat
sns.heatmap(conf_mat_df, annot=True, annot_kws={"size": 30}, fmt='d',cmap="seismic", ax = axes )
axes.set_title('Confusion Matrix')
plt.show()
```

# 2. Multinomial Naive Bayes Algorithm

```python
NB= MultinomialNB()
NB.fit(x_train,y_train)
```

```python
# Predict on the test set and calculate accuracy
NB_pred=NB.predict(x_test)
score=accuracy_score(y_test,NB_pred)
```

```python
print('Multinomial Naive Bayes')
print(classification_report(y_test,NB_pred))
print(f'Accuracy: {round(score*100,2)}%')
conf_mat = confusion_matrix(y_test, NB_pred)
axes = plt.axes()
conf_mat_df = conf_mat
sns.heatmap(conf_mat_df, annot=True, annot_kws={"size": 30}, fmt='d',cmap="seismic", ax = axes )
axes.set_title('Confusion Matrix')
plt.show()
```

# 3. KNN Algorithm

```python
KNN=KNeighborsClassifier(n_neighbors=6)
KNN.fit(x_train, y_train)
```

```python
# Predict on the test set and calculate accuracy
KNN_pred=KNN.predict(x_test)
score=accuracy_score(y_test,KNN_pred)
```

```python
print('K-Nearest Neighbour')
print(classification_report(y_test,KNN_pred))
print(f'Accuracy: {round(score*100,2)}%')
conf_mat = confusion_matrix(y_test, KNN_pred)
axes = plt.axes()
conf_mat_df = conf_mat
sns.heatmap(conf_mat_df, annot=True, annot_kws={"size": 30}, fmt='d',cmap="seismic", ax = axes )
axes.set_title('Confusion Matrix')
plt.show()
```

# 4. Decision Tree Classifier Algorithm

```python
DTC = DecisionTreeClassifier(max_depth=10)
DTC.fit(x_train,y_train)
```

```python
# Predict on the test set and calculate accuracy
DTC_pred=DTC.predict(x_test)
score=accuracy_score(y_test,DTC_pred)
```

```python
print('K-Nearest Neighbour')
print(classification_report(y_test,DTC_pred))
print(f'Accuracy: {round(score*100,2)}%')
conf_mat = confusion_matrix(y_test, DTC_pred)
axes = plt.axes()
conf_mat_df = conf_mat
sns.heatmap(conf_mat_df, annot=True, annot_kws={"size": 30}, fmt='d',cmap="seismic", ax = axes )
axes.set_title('Confusion Matrix')
plt.show()
```
